Gradient Descent

```
alpha=0.01

a=int(input("Enter intial value for a"))
b=int(input("Enter intial value for b"))

epoch=10000

x=[1,2,3,4,5]
y=[10,40,50,78,83]
c=[]
m_a=[]
m_b=[]


def cost(y,h):
  t=[((h[i]-y[i])**2) for i in range(len(y))]   
  cost=sum(t)/(2*len(y))
  return cost

def d_b_cost(y,h):
  t=[((h[i]-y[i])) for i in range(len(y))]   
  d_b_cost=sum(t)/(len(y))
  return d_b_cost


def d_a_cost(y,h,x):
  t=[((h[i]-y[i])*(x[i])) for i in range(len(y))]   
  d_a_cost=sum(t)/(len(y))
  return d_a_cost

for i in range(epoch):
  h=[((a*k)+b) for k in x] 
  c.append(cost(y,h))
  m_a.append(a)
  m_b.append(b)
  t_a=d_a_cost(y,h,x)
  t_b=d_b_cost(y,h)
  a=a-(alpha)*(t_a)
  b=b-(alpha)*(t_b)

print(a,b)
```
